```python
inputs = np.stack(get_data('SAM',200,['qvi','ts','huss','tas'],['qvi','ts','huss','tas'],
                           scaling_method = 'minmax'))

outputs = np.stack(get_data('SAM',200,['pracc','clt'],['pracc','clt'],scaling_method = 'none'))

outputs = np.moveaxis(outputs,0,3)
inputs  = np.moveaxis(inputs,0,3)
```

```python
outputs[1:,:,:,0] = outputs[1:,:,:,0]-outputs[:-1,:,:,0]
inputs = inputs[1:,:,:,:]
outputs = outputs[1:,:,:,:]
```

```python
# lonlat= (post_process.lon_lat()).reshape(1,22,192,1)

outputs = outputs[1:,:,:,:]-outputs[:-1,:,:,:]
inputs = inputs[1:,:,:,:]
# outputs*=lonlat
```

```python
def rand(nn):
    import random
    random.seed(9001)
    x = [random.randint(1,100) for i in range(0,nn)]
    return np.array(x)
def chunk(datain,window):
    nt,nlat,nlon,nfeature = datain.shape
    nchunk = int(nt/window)
    dtin = datain[:nchunk*window ,:,:,:]
    dtin = np.reshape(dtin,(nchunk,window,nlat,nlon,nfeature))
    return dtin
def reshape_data(data):
    nt,window,nlat,nlon,nfeature = data.shape
    
    dtin = np.moveaxis(data,1,3)
    dtin = np.reshape(dtin,(nlat*nlon*nt,window,nfeature))
    return dtin

timeframe=3


x_all =  chunk(inputs,timeframe)
y_all =  chunk(outputs,timeframe)
mask =  rand(y_all.shape[0]) < 70
in_train,in_test = x_all[mask,:,:,:], x_all[~mask,:,:,:]
out_train,out_test = y_all[mask,:,:,:], y_all[~mask,:,:,:]

x_train = reshape_data(in_train)
x_test = reshape_data(in_test)
y_train = reshape_data(out_train)
y_test = reshape_data(out_test)


```

```python
units = [200,128,128,64]
l1 = 0e-7

initializer = tf.keras.initializers.GlorotNormal()
batch_size = 256
epochs = 30
activation = tf.nn.relu
output_shape = 1
inshapeX = x_train[0,:,:].shape
val_split = 0.2


input_layer = Input(shape=(inshapeX), name='inputs')
xs = Flatten()(input_layer)
xs =  Dense(units=units[0],activation = activation,kernel_initializer=initializer,
            name = 'dense1')(xs)

xs =  Dense(units=units[1],activation = activation,kernel_initializer=initializer,
           name = 'dense2')(xs)

xs =  Dense(units=units[2],activation = activation,kernel_initializer=initializer,
           name = 'dense3')(xs)

xs =  Dense(units=units[3],activation = activation,kernel_initializer=initializer,
            name = 'dense4')(xs)

y_predict =  Dense(units=output_shape,activation = 'sigmoid',name = 'prediction')(xs)

print (y_predict.shape)


nn_precip = Model(inputs = [input_layer],
              outputs = [y_predict]) 

opt = keras.optimizers.Adam(lr=0.0001)

nn_precip.compile(optimizer = opt,loss = 'mse')

earlyStopping=keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, verbose=0, mode='auto')



hist = nn_precip.fit({'inputs' : x_train[:,:,:]},
                  {'prediction': y_train[:,2,1]},
                   epochs = epochs,  validation_split = val_split,callbacks= [earlyStopping],
                     batch_size = batch_size,verbose = 1, shuffle=True)
```

```python
def plot_colormap(data_ax1,data_ax2,data_color,num_bin):
    mx1 = np.max(data_ax1)
    mn1 = np.min(data_ax1)
    binsize1 = (mx1 - mn1)/num_bin
    
    mx2 = np.max(data_ax2)
    mn2 = np.min(data_ax2)
    binsize2 = (mx2 - mn2)/num_bin
    bin_ax1 = np.linspace(mn1,mx1,num_bin)
    bin_ax2 = np.linspace(mn2,mx2,num_bin)
    
    precip_mean = np.zeros((num_bin,num_bin))
    precip_std = np.zeros((num_bin,num_bin))
    error_mean = np.zeros((num_bin,num_bin))
    error_std = np.zeros((num_bin,num_bin))
    numpoint = np.zeros((num_bin,num_bin))
    
    precip_mean.fill(np.nan)
    precip_std.fill(np.nan)
    error_mean.fill(np.nan)
    error_std.fill(np.nan)
    
    for i in tqdm (range(num_bin-1)):
#         thi1 = mn1 + i*binsize1
#         thi2 = mn1 + (i+1)*binsize1
        for j in range(num_bin-1):
            indecies=np.where(np.logical_and(np.logical_and(data_ax1>=bin_ax1[i],data_ax1<bin_ax1[i+1]),
                                    np.logical_and(data_ax2>=bin_ax2[j],data_ax2<bin_ax2[j+1])))[0]
            if len(indecies)>5:
                numpoint[j,i] = len(indecies)

                precip_mean[j,i] = np.mean(data_color[indecies])
                precip_std[j,i] = np.std(data_color[indecies])
                
                error_mean[j,i] = np.mean(data_color[indecies] - data_ax2[indecies])
                error_std[j,i] = np.std(data_color[indecies]  - data_ax2[indecies])
    return precip_mean,precip_std,error_mean,error_std,numpoint
                
                
    
```

```python

def Compute_R2_lonlat(data,true,model):
    nt,t_wind, lat, lon, features = data.shape
    R2matrix = np.zeros((lat,lon))
    R2matrix[:] = np.nan
    for yy in tqdm (range(lat)):
        for xx in range(lon):
            test = data[:,:,yy,xx,:]
            true_i = true[:,:,yy,xx,:]
            
            index = np.where(true_i[:,-1,0]>.05)[0]
            if len(index)>10:
                test = (test[index,:]).reshape(-1,t_wind,features)
                true_i = true_i[index,:]
                predict = model.predict(test)
                R2matrix[yy,xx] = r2_score(predict[:,0],true_i[:,-1,0])
    return R2matrix

```
